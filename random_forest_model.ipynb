```python
def create_subsets(rf_meta, nullvars, rf_taxa):
    base_data = rf_meta.drop(columns=nullvars, errors='ignore')
    with_taxa_data = pd.concat([base_data, rf_taxa], axis=1)
    return {"Base": base_data, "WithTaxa": with_taxa_data}
```

```python
def preprocess_data(data, varsnot2norm):
    scaler = StandardScaler()
    continuous_vars = data.columns.difference(varsnot2norm)

    data_scaled = data.copy()
    data_scaled[continuous_vars] = scaler.fit_transform(data[continuous_vars])
    
    return data_scaled
```
